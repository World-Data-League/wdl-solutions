### 3. Introduction

According to the [World Health Organization](https://www.who.int/health-topics/air-pollution#tab=tab_1), ‚Äú**air pollution** is contamination of the indoor or outdoor environment by any chemical, physical or biological agent that modifies the natural characteristics of the atmosphere‚Äù. Household combustion devices and motor vehicles are common sources of air pollution, which might be two main causes for the bad air quality in Stadhouderskade. 

It is estimated that 9 out of 10 people worldwide live in places where air quality exceeds WHO guideline limits. As a consequence of high levels of air pollution, people might risk getting diseases like respiratory infections, lung cancer and heart disease. The most health-harmful pollutants are PM2.5 particles that penetrate deep into lung passageways.

This is why Sustainable Development Goal 11 is so important.

#### 3.1) Air Pollutants in Stadhouderskade

The street‚Äôs air quality is measured hourly according to eight components:
-	Nitrogen Monofluoride: FN
-	Particles of a diameter of 2.5 Œºm or less (Particulate Matter 2.5): PM2.5
-	Particles of a diameter of 10 Œºm or less (Particulate Matter 10): PM10
-	Nitric Oxide: NO
-	Nitrogen Dioxide: NO2
-	Cycloheptatriene: C7H8
-	Benzene: C6H6
-	Ethylbenzene: C8H10

Their concentration in the air is calculated by a sensor as micrograms per cubic meter (¬µg/m¬≥).

Now, we will give more details into some of these air pollutants:

**a) Particulate Matter 2.5 (PM2.5) and Particulate Matter 10 (PM10):**

Airborne particulate matter (PM) is not a single pollutant, but rather a mixture of many chemical species. Fine particulate matter is defined as particles that are 2.5 microns or less in diameter (PM2.5). Those with a diameter of 10 microns or less (PM10) are inhalable into the lungs and can induce adverse health effects.

Emissions from combustion of gasoline, oil, diesel fuel or wood produce much of the PM2.5 pollution found in outdoor air, as well as a significant proportion of PM10.

Several adverse health impacts have been associated with exposure to particle matter. PM2.5 has been associated with heart or lung problems, bronchitis, and asthma attacks. PM10 has been associated primarily with asthma and chronic obstructive pulmonary disease (COPD).

*Source: [California Air Resources Board](https://ww2.arb.ca.gov/resources/inhalable-particulate-matter-and-health)*

**b) Nitric Oxide (NO) and Nitrogen Dioxide (NO2):**

Nitric oxide and nitrogen dioxide are referred to as oxides of nitrogen (NOx). When nitrogen is released during fuel combustion it combines with oxygen atoms to create nitric oxide (NO). This further combines with oxygen to create nitrogen dioxide (NO2).

In areas of high motor vehicle traffic, such as in large cities, the amount of nitrogen oxides emitted into the atmosphere as air pollution can be significant.

Nitric oxide is not considered to be hazardous to health at typical ambient concentrations, although nitrogen dioxide is an irritant gas, which at high concentrations causes inflammation of the airways.

*Source: [Icopal Noxite](http://www.icopal-noxite.co.uk/nox-problem/nox-pollution.aspx)*

**c) Cycloheptatriene (C7H8), Benzene (C6H6) and Ethylbenzene (C8H10):**

Cycloheptatriene, benzene and ethylbenzene are toxic volatile organic compounds, which are atmospheric pollutants released into the environment in urban settings.

These can be found on natural form in crude oil, diesel, and gasoline. Also, some of these compounds can exist in synthetic materials, consumer products, fuel additives, among others.

They can enter organisms through various routes, the most common being inhalation, representing a threat to human health.

*Source: Montero-Montoya, R., L√≥pez-Vargas, R. and Arellano-Aguilar, O. (2018). Volatile Organic Compounds in Air: Sources, Distribution, Exposure and Associated Illnesses in Children. Annals of Global Health, 84(2), pp. 225‚Äì238. DOI: http://doi.org/10.29024/aogh.910*

#### 3.2) Air Quality Standards

Stadhouderskade is a street in Amsterdam, the Netherlands‚Äô capital. This country is a part of the European Union, which supported a project called CITEAIR. This project introduced the **Common Air Quality Index (CAQI)**.

This index can be used to easily compare air quality across the EU and warn citizens of the dangers of high levels of pollution. In this project, we will use the hourly 'roadside' index. "Hourly" because we have hourly data for Stadhouderskade and "roadside" because Stadhouderskade is considered as a city street with a lot of traffic. The two mandatory measurements for the 'roadside' CAQI are NO2 and PM10, PM2.5 being optional.

The CAQI is defined as the sub-index that represents the worst quality among those three components (i.e., NO2, PM10 and PM2.5).

The table below shows the ranges for each pollutant to apply CAQI. Note that Ozone (O3) is in the table, but it is not used on the 'roadside' index.

**c) Build the Final Dataset:**

Start by merging the air pollutants data with the weather data: 

```python
# Save old index for later
idx = table.index

# Set new index with date only
table.set_index(table.index.date, inplace=True)

# Merge by indexes
table = pd.merge(table, 
                 weather, 
                 how='left',
                 left_index=True, 
                 right_index=True)

# Set old index 
table.set_index(idx, inplace=True)
```

We have a main difference between these two data sources: the air pollution dataset has **hourly data** and the weather dataset has **daily data**. 

We could choose any structure, but we decided to continue with hourly data.

Thus, for the temperature data we want to assign the minimum temperature to the first and last 6 hours of the day, and the maximum temperature for the 12 hours in the middle. If these values are missing, we assign the average temperature for the whole 24 hours.

```python
# Create 'hour' column
table['hour'] = table.index.hour

# First and last 6 hours -> min temp
idx_temp_min = table.loc[(table['hour']<=5) | (table['hour']>=18)].index

# Middle 12 hours -> max temp
idx_temp_max = table.loc[(table['hour']>5) | (table['hour']<18)].index

# Assign temperature values according to the rule defined above
for i in table.index:

    if i in idx_temp_min and table.at[i, 'tmin']!=np.nan:
        table.at[i, 'temp'] = table.at[i, 'tmin']

    elif i in idx_temp_max and table.at[i, 'tmax']!=np.nan:
        table.at[i, 'temp'] = table.at[i, 'tmax']

    else:
        table.at[i, 'temp'] = table.at[i, 'tavg']
```

```python
# Drop unnecessary columns
table.drop(['tavg', 'tmin', 'tmax', 'hour'], axis=1, inplace=True)

# Check the first day on the dataset
table.head(24)
```

However, according to the index, **PM2.5** and **PM10** have some extremely high levels of pollution that are also isolated data points in their respective box plots. Thus, these two might have outliers.

Therefore, we want to deepen the analysis because **C6H6**, **C7H8**, **C8H10**, **PM2.5** and **PM10**  might have extreme outliers. In this case, we will smooth the series with two bounds, upper and lower. These bounds depend on the mean absolute error, obtained from the difference between real data and smoothed values. The values above (or below) this interval are considered as outliers.

We have defined two functions for the smoothing analysis. The first one helps us calculates the bounds. The second smooths and plots the series.

```python
def mean_absolute_error(y_true, y_pred):
    return np.mean(np.abs(y_true - y_pred))

def plot_moving_average(series, window, plot_intervals=False, scale=1.96):
    rolling_mean = series.rolling(window=window).mean()
    
    plt.figure(figsize=(17,8))
    plt.suptitle(f'Moving Average ({series.name})', fontsize=30)
    plt.title(f'Window Size = {window} hours', fontsize=15)
    plt.plot(rolling_mean, 'g', label='Rolling mean trend')
    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        deviation = np.std(series[window:] - rolling_mean[window:])
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        plt.plot(upper_bound, 'r--', label='Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')
            
    plt.plot(series[window:], label='Actual values')
    plt.legend(loc='best')

    plt.show()
```

We will show the PM2.5 values as an example. We chose to visualize a year from June 2018 to May 2019, as the highest concentration of PM2.5 was hit in this interval (January 2019).

#### 7.1) Decomposition and Autocorrelation

Stationarity is essential for a time series to be fed to an ARIMA model. A time series is stationary when mean, variance and co-variance are constant over time.

We also have to guarantee the series is not autocorrelated, meaning the independant variables should be uncorrelated with the precedant time lags.

If a series is not stationarity or is autocorrelated, we have ways to deal with it. We can transform the time series with differentiation, so we subtract every value by its precedent.

The *dickey_fuller_plot* function below analyzes a time series' stationarity and autocorrelation. It does it with the Augmented Dickey-Fuller test. This allows us to detect which series are ready to use (stationary and not autocorrelated). And also the ones that need to be transformed.

The function evaluates the null hypothesis that there is autocorrelation, so if p-value is below 0.05 we consider that there is no correlation between the moments. Otherwise, Dickey-Fuller recommends to transform the dataset. Furthermore, we also plot autocorrelation and partial autocorrelation.

```python
def dickey_fuller_plot(data, pollutant, lags=8):
    series = data[pollutant]
    series_dif = (series - series.shift(1))[1:]

    # https://www.statsmodels.org/devel/generated/statsmodels.tsa.stattools.adfuller.html
    p_value = sm.tsa.stattools.adfuller(series)[1]
    p_value_dif = sm.tsa.stattools.adfuller(series_dif)[1]

    fig = plt.figure(figsize=(12, 7))
    layout = (3,2)
    ts_ax = plt.subplot2grid(layout, (0,0), colspan=2)
    acf_ax = plt.subplot2grid(layout, (1,0))
    pacf_ax = plt.subplot2grid(layout, (1,1))

    acf_ax_dif = plt.subplot2grid(layout, (2,0))
    pacf_ax_dif = plt.subplot2grid(layout, (2,1))
    
    series.plot(ax=ts_ax)
    ts_ax.set_title(f'Time Series Analysis Plots: {pollutant}\n Dickey-Fuller: p={p_value:.5f}\n DF diff: p={p_value_dif:.5f}')
    smt.graphics.plot_acf(series, lags=lags, ax=acf_ax)
    smt.graphics.plot_pacf(series, lags=lags, ax=pacf_ax)
    
    smt.graphics.plot_acf(series_dif, lags=lags, ax=acf_ax_dif)
    smt.graphics.plot_pacf(series_dif, lags=lags, ax=pacf_ax_dif)

    plt.tight_layout()
```

```python
# Display Dickey-Fuller for 'PM25' as an example
dickey_fuller_plot(final_table, 'PM25', 24)
```

The outputs above show us the Dickey-Fuller p-value and the correlogram and partial correlogram for the original series (first + second) and for the differenciated series (third + fourth).

The autocorrelation plots consider internal association between observations. The closer to 1 or -1, the highest correlation between lags. 

The partial correlogram is a summary of the relationship between an observation in a time series with other prior observation, removing the relationship of observations contained between them.

We checked all variables and concluded all of them are stationary (all p-values < 0.05), so there is no need to apply transformations. The output above shows the results for PM2.5, as an example.

We will, finally, start building the 3 ARIMA models. 

ARIMA stands for:
- **AR - Autoregressive:** it is a regression of the variable against itself with a lag p of observations; in an autoregression model, we forecast using a linear combination of past values of the variable
- **I - Integrated:** the use of differencing of raw observations to make the time series stationary
- **MA - Moving Average:** a moving average term in a time series model is a past error (multiplied by a coefficient); uses the dependency between an observation and a residual error from a moving average model applied to
lagged observations

Simply put, the autoregression of the residual errors is the moving average model.

The *optimize_ARIMA* function below finds the model with lowest AIC (Akaike Information Critera). It tests several models, changing the order parameters p,d and q. These determine the number of lags for the Autorregresive and Moving Average models (parameters p and q, respectively). d represents the integration parameter, which indicates the number of differentation series needed to reach stationarity.


```python
# Set initial values and some bounds
p = range(0, 2)
d = 1
q = range(0, 2)

# Create a list with all possible combinations of parameters
parameters = product(p, q)
parameters_list = list(parameters)
parameters_list
```

```python
def optimize_ARIMA(parameters_list, endog, exog, d=1):
    """
    We enter a series and the function returns a dataframe with parameters and corresponding AIC
    
    parameters_list - list with (p, q) tuples
    p - AR time laps
    d - integration order
    q - MA time laps

    endog - observed time-series process, y
    
    exog - array of exogenous regressors
    """
    results = []
    best_aic = float('inf')
    
    for param in tqdm_notebook(parameters_list):
        try: 
            model = ARIMA(endog=endog, exog=exog, order=(param[0], d, param[1]))
            model_fit = model.fit(disp=-1)
        except:
            continue
            
        aic = model_fit.aic
        
        # Save best model, AIC and parameters
        if aic < best_aic:
            best_model = model_fit
            best_aic = aic
            best_param = param
        results.append([param, model_fit.aic])

    result_table = pd.DataFrame(results)
    result_table.columns = ['parameters', 'aic']
    
    # Sort in ascending order, lower AIC is better
    result_table = result_table.sort_values(by='aic', ascending=True).reset_index(drop=True)
    
    return result_table
```

The next function represents the forecasts made from our ARIMA models. It also includes the error calculation and residuals analysis. Predictions are represented as the green line in the chart.

We defined three error indicators:
- MAPE (Mean Absolute Percentage Error);
- MAE (Mean Absolute Error);
- RMSE (Root Mean Squared Error).


```python
def forecast_resid_error(train_data, valid_data, model_fit, exog, steps=23):
    # Forecast
    fc, se, conf = model_fit.forecast(steps, exog, alpha=0.05) # 95% conf
    train_series = train_data
    valid_series = valid_data
    series = pd.concat([train_series, valid_series])
    
    # Error calculation
    mape = np.mean(np.abs(fc - valid_data)/np.abs(valid_data))
    mae = np.mean(np.abs(fc - valid_data))    
    rmse = np.mean((fc - valid_data)**2)**.5  

    print({'mape': round(mape, 3), 'mae': round(mae, 3), 'rmse': round(rmse, 3)})

    # Lower and upper limits
    fc_series = pd.Series(fc, index=valid_series.index)
    lower_series = pd.Series(conf[:, 0], index=valid_series.index)
    upper_series = pd.Series(conf[:, 1], index=valid_series.index)

    # Plot
    plt.figure(figsize=(12,5), dpi=100)

    plt.plot(train_data.loc['2022-02-01':], label='training') # Check only a subset to better visualize data
    plt.plot(valid_data, label='actual')
    plt.plot(fc_series, label='forecast')

    plt.fill_between(lower_series.index, 
                     lower_series, upper_series, 
                     color='k', alpha=.15)
    
    plt.title('Forecast vs Actuals')
    plt.legend(loc='upper left', fontsize=8)

    plt.show()

    # Plot residual errors
    residuals = pd.DataFrame(model_fit.resid)

    fig, ax = plt.subplots(1,2)

    residuals.plot(title="Residuals", ax=ax[0])
    residuals.plot(kind='kde', title='Density', ax=ax[1])

    plt.show()

    return fc
```

```python
# Forecast for NO2
fc_NO2 = forecast_resid_error(train['NO2'], valid['NO2'], model_NO2_fit, valid.drop(['NO2', 'CAQI'], axis=1))
```

```python
# Forecast for PM10
fc_PM10 = forecast_resid_error(train['PM10'], valid['PM10'], model_PM10_fit, valid.drop(['PM10', 'CAQI'], axis=1))
```

```python
# Forecast for PM2.5
fc_PM25 = forecast_resid_error(train['PM25'], valid['PM25'], model_PM25_fit, valid.drop(['PM25', 'CAQI'], axis=1))
```

The quality on the forecasts could be improved. We could make many more tests by adding/removing variables or even by using only lags to make predictions (univariate models), but we decided to continue with these 3 models.

In addition, residuals are the difference between fitted and observed values. It is very important to analyze them, because they can give us information about how the model is fitting. There are two main properties that residuals have to achieve:
- residuals must be uncorrelated between them; if not, it means there is still information in the residuals that is not being used
- residuals must have zero mean

If these two conditions are not met, the model has not adequately represented our data.

The table below shows us the predictions for the next 5 hours.



```python
# Show 5 hour forecast for NO2
frcst_NO2 = pd.DataFrame({'frcst_NO2':fc_NO2,
                          'valid':valid['NO2'],
                          'dif':fc_NO2-valid['NO2'],
                          'dif%':fc_NO2/valid['NO2']})

frcst_NO2.head()
```

```python
# Show 5 hour forecast for PM10
frcst_PM10 = pd.DataFrame({'frcst_PM10':fc_PM10,
                          'valid':valid['PM10'],
                          'dif':fc_PM10-valid['PM10'],
                          'dif%':fc_PM10/valid['PM10']})

frcst_PM10.head()
```

```python
# Show 5 hour forecast for PM2.5
frcst_PM25 = pd.DataFrame({'frcst_PM25':fc_PM25,
                          'valid':valid['PM25'],
                          'dif':fc_PM25-valid['PM25'],
                          'dif%':fc_PM25/valid['PM25']})

frcst_PM25.head()
```

We will end the analysis by merging the predictions from our 3 ARIMA models. Using this merged dataset, we can generate the **Common Air Quality Index (CAQI) pollution levels** for each hour of predictions, because we will have predictions for the concentrations of NO2, PM10 and PM2.5, which are the values needed to generate CAQI.

```python
# Merge forecasts
frcst = pd.merge(frcst_NO2, 
                 frcst_PM10, 
                 how='left',
                 left_index=True, 
                 right_index=True)

frcst = pd.merge(frcst, 
                 frcst_PM25, 
                 how='left',
                 left_index=True, 
                 right_index=True)

frcst = frcst[['frcst_NO2', 'frcst_PM10', 'frcst_PM25']]
```

```python
# CAQI_pred: like it was said in section 3.2, CAQI is "the sub-index that represents the worst quality among those three components"
frcst['CAQI_pred'] = frcst[['frcst_NO2', 'frcst_PM10', 'frcst_PM25']].max(axis=1)

frcst['CAQI_pred_level'] = ''

# CAQI_pred_level: Index according to the 2 first columns on the table from 3.2
frcst.loc[frcst['CAQI_pred'] <= 25, 'CAQI_pred_level'] = '1 - very low'
frcst.loc[frcst['CAQI_pred'].between(25, 50), 'CAQI_pred_level'] = '2 - low'
frcst.loc[frcst['CAQI_pred'].between(50, 75), 'CAQI_pred_level'] = '3 - medium'
frcst.loc[frcst['CAQI_pred'].between(75, 100), 'CAQI_pred_level'] = '4 - high'
frcst.loc[frcst['CAQI_pred'] > 100, 'CAQI_pred_level'] = '5 - very high'

frcst[['CAQI_pred', 'CAQI_pred_level']].head()
```

Using the concentration values forecasted by our models and the CAQI levels generated from those forecasts, we have the information needed to make an "Air Pollution Dashboard" for Stadhouderskade: 
- NO2, PM10 and PM2.5 concentration predictions (in ¬µg/m¬≥)
- CAQI levels with a 1-5 scale (1 being 'very low pollution' and 5 being 'very high pollution')

#### 8.2) Solutions and Implementations

We will now propose some possible solutions for the issue of air pollution in Stadhouderskade. These are mainly directed to two groups: the people living in Amsterdam (specially the ones living in Stadhouderskade, or frequently passing by it) and Amsterdam's city council. We believe these are the people which can more likely influence positive change for this part of Amsterdam.

Therefore, in order to help the Green Mile project in transforming Stadhouderskade into a place where people stay, instead of just transiting through, we suggest the following measures:
- Encourage people to **check a dashboard** like the one in section 8.1. This could be similar to checking weather predictions for the next couple of days. Citizens would be able to check the air pollution predictions. If the air pollution levels were 'high' or 'very high', some restrictions could be implemented (e.g if air quality was very poor, restrictions could be applied to vulnerable people with respiratory diseases - maybe they should avoid passing on the street). This could reduce the number of people exposed to overly polluted air.

  Also, if the city council had access to this information, maybe the Green Mile project would be able to raise awareness among policymakers.

- **Restrictions to car traffic** are also important. Like we saw in section 3, cars are one of the main sources of air pollution. Thus, we suggest the creation of a "Low Emission Zone" where only cars registered after 2011 can circulate in Stadhouderskade. This would follow the *Euro 5* emissions standard from EU (source: [Euro 1 to Euro 6 guide](https://www.rac.co.uk/drive/advice/emissions/euro-emissions-standards/)). This could be applied more strictly if the air pollution level was predicted to be 'high' or 'very high'.

  This measure would improve the public space and lead to better circulation conditions for residents.
- If we want the people to be out on the street, Stadhouderskade should become more pedestrian friendly. There could be a better infrastructure for the citizens to walk, run, ride a bike, play sports, or even walk their dog, play outside with children and practise yoga. We believe the existence of outdoor activities could be key in making people want to stay longer on the street.
  
  Essentially, the city council could try to **optimize and add public/green spaces** in the vicinity of the street. In fact, this street has a lot of parking lots that could be relocated in order to create more green areas, gardens or parks. This would make the cars stay further away, inviting pedestrians to stay.

  This is backed up by the analysis we did in section 4.1, which showed that near to Stadhouderskade there are only 2 city parks, no running routes, some sports providers (although there are regions in the city with more) and only one
sports park. These types of sports infrastructure could be added to the vicinity of Stadhouderskade to invite people to practise sports and stay there longer.
- Put a couple of **outdoor interactive banners** along the street displaying the amount of air pollutants that were emitted over a certain period where people could filter and visualize these amounts in real time. Also, the banners could display how much pollution could be avoided if more people had decided to pass the street by walking or riding a bike, instead of driving a car.

  This could clearly inform people about their environmental impacts, making them more conscious about air pollution. Then, many could start thinking: "Do I really need to drive my car today? Can I ride a bike or walk to where I am headed?"
- Finally, some **marketing campaigns** could be done to educate citizens, using flyers, outdoor posters, and most importantly, social media. This would explain all the measures to the citizens, and it should also explain the main sources of air pollution (and the main solutions!).

#### 8.3) Future Work

- More data could be collected by the sensor(s) to be added to the time series. For example, data about the concentration of other air pollutants (e.g. ground-level ozone, methane, carbon monoxide), car traffic (e.g. average number of cars passing on Stadhouderskade per hour) and noise levels.
- A predictive model could be created for each one of the 7 districts in Amsterdam (Centre, West, Oost, Noord, Zuid, Nieuw-West and Zuidoost), thus turning this project into more of a high-level analysis from which all the citizens of Amsterdam could benefit from.
- We could apply a VARMAX model with 3 endogenous variables (NO2, PM10, PM2.5), instead of building one ARIMA model for each pollutant.
- Also, our time series modeling could be turned into a classification problem if the dependent variable was turned into classes (for example using the classes on the Common Air Quality Index from EU). Nonetheless, we would apply algorithms to perform time series classification that are probably more complex than ARIMA.
